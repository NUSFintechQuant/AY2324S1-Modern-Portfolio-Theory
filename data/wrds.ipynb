```python
# idx_daily isn't on the wrds webpage, but some digging via `conn.get_tables(library='comp_na_daily_all)` yields other tables such as idx_daily.
# gvkeyx is a unique code compustat assigns to a security; 000003 is S&P
# reference: https://wrds-www.wharton.upenn.edu/data-dictionary/comp_na_daily_all/idx_daily/ for variable descriptions
# can just use `prccddiv` since it takes dividends returns into account

comp_snp_historical_daily = conn.raw_sql("""
select *
from comp_na_daily_all.idx_daily
where gvkeyx = '000003'
""")
comp_snp_historical_daily
```

```python
# we use us bond ETFs since bonds are issued at intervals -> no one ticker to track
# bond ETF below used by Z. Cazalet and T. Roncalli, “Facts and Fantasies About Factor Investing.” Rochester, NY, Oct. 31, 2014. doi: 10.2139/ssrn.2524547.
# https://papers.ssrn.com/sol3/papers.cfm?abstract_id=2524547
# ref: https://etfdb.com/etf/TLT/#etf-ticker-profile
# note: secd: https://wrds-www.wharton.upenn.edu/data-dictionary/comp_na_daily_all/secd/
# note: prccd refers to price - close - daily

comp_us_bonds_historical_daily = conn.raw_sql("""
select datadate, conm, tic, prccd
from comp_na_daily_all.secd
where tic = 'TLT'
""")
comp_us_bonds_historical_daily
```

```python
comp_snp_historical_daily.to_csv('./snp500.csv')
comp_us_bonds_historical_daily.to_csv('./us_bonds.csv')
```
